```python
import os
from pathlib import Path
base_dir = Path()

os.listdir(base_dir)
```

```python
class_names = os.listdir(os.path.join(base_dir ,'data','train'))
class_names
```

```python
import matplotlib.pyplot as plt
import cv2
class_names = os.listdir(os.path.join(base_dir ,'data','train'))
cls_dict = {}

plt.figure(figsize=(7, 12))

for i, class_name in enumerate(class_names):
    cls_dict[class_name] = os.listdir(os.path.join(base_dir,'data','train', class_name))[:5]
    
    for j in range(1, 6):
        img_path = os.path.join(base_dir, 'data','train',class_name, cls_dict[class_name][j-1])
        img = cv2.imread(img_path)
        
        # Calculate the subplot index
        subplot_index = i * 5 + j
        
        # Plot the image in the corresponding subplot
        plt.subplot(7, 5, subplot_index)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
        plt.title(class_name)
        plt.axis('off')

plt.show()
```
